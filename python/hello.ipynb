```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
data=load_iris()
X,y=data["data"],data["target"]
X[:10]
```

```python
sns.set_theme(style="white",rc={"axes.facecolor":(0, 0, 0, 0)})
#Create thedata
x=X.ravel()
labels= ["sepallength", "sepalwidth","petallength", "petalwidth"]
g=np.tile(labels,len(X))
df=pd.DataFrame(dict(x=x,g=g))
#InitializetheFacetGridobject
pal=sns.cubehelix_palette(10,rot=-.25,light=.7)
g=sns.FacetGrid(df,row="g",hue="g",aspect=10,height=1.5,palette=pal)
#Drawthe densities
g.map(sns.kdeplot, "x",bw_adjust=.5,clip_on=False,fill=True,alpha=1,linewidth=1.5)
g.map(sns.kdeplot, "x",clip_on=False,color="w",lw=2,bw_adjust=.5)
#Addreferenceline
g.refline(y=0,linewidth=2,linestyle="-",color=None,clip_on=False)
```
